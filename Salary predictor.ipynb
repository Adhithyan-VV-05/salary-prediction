```python
#Step 1 : import pandas library
import pandas as pd
```

```python
#Step 2 : Read file
salary=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv")
```

```python
salary
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Experience Years']
  ys = series['Salary']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = salary.sort_values('Experience Years', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Experience Years')
_ = plt.ylabel('Salary')
```

```python
from matplotlib import pyplot as plt
salary.plot(kind='scatter', x='Experience Years', y='Salary', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)

```

```python
#Step 3 : define targets (y) and features (x)
y=salary["Salary"]
x=salary[["Experience Years"]]
```

```python
salary.columns
```

```python
#Step 4 : splitting data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)            # no changes neeeded
```

```python
#Step 5 : select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# Step 6 : Train model ( fit model ) <---- ML Happens
model.fit(x_train,y_train)                                                       # no changes needed
```

```python
print(type(x_test))
```

```python
 #Step 7 : Prediction
 y_pred=model.predict(x_test)
```

```python
# Step 8 : Accuracy
from sklearn.metrics import mean_absolute_percentage_error                       # no changes needed
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
